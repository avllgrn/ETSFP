Python identifica distintos tipos de datos. Unos pueden ser numéricos (con o sin parte decimal) o cadenas de caracteres:

```python
a = 1
b = 2.3
c = '@'

print('a =',a,'\t\ttype(a) =',type(a))
print('b =',b,'\ttype(b) =',type(b))
print('c =',c,'\t\ttype(c) =',type(c))
```

En Python, puede solicitarse al usuario información; pero, ésta siempre se identifica como cadena de caracteres:

```python
a = input('Ingresa un número sin parte decimal ')
b = input('Ingresa un número con parte decimal ')
c = input('Ingresa un caracter ')

print('a =',a,'\t\ttype(a) =',type(a))
print('b =',b,'\ttype(b) =',type(b))
print('c =',c,'\t\ttype(c) =',type(c))
```

Si se desea solicitar información de distinto tipo, debe convertirse al tipo de dato que se necesite:

```python
a = int(input('Ingresa un número sin parte decimal '))
b = float(input('Ingresa un número con parte decimal '))
c = input('Ingresa un caracter ')

print('a =',a,'\t\ttype(a) =',type(a))
print('b =',b,'\ttype(b) =',type(b))
print('c =',c,'\t\ttype(c) =',type(c))
```

En Python, pueden realizarse cálculos aritméticos, utilizando operadores.

```python
a = int(input('Ingresa un número sin parte decimal '))
b = int(input('Ingresa un número sin parte decimal '))

c = a + b
d = a - b
e = a * b
f = a / b
g = a // b
h = a % b
i = a ** b

print('a      =',a,'\ttype(a) =',type(a))
print('b      =',b,'\ttype(b) =',type(b))
print('a + b  =',c,'\ttype(a + b) =',type(c))
print('a - b  =',d,'\ttype(a - b) =',type(d))
print('a * b  =',e,'\ttype(a * b) =',type(e))
print('a / b  =',f,'\ttype(a / b) =',type(f))
print('a // b =',g,'\ttype(a // b) =',type(g))
print('a % b  =',h,'\ttype(a % b) =',type(h))
print('a ** b  =',i,'\ttype(a ** b) =',type(i))
```

En Python, existe una gran cantidad de funciones matemáticas que nos permiten realizar cálculos de mayor grado de dificultad.

```python
import math

grados = float(input('Ingresa un ángulo en grados '))

radianes = grados * math.pi / 180

seno = math.sin(radianes)
coseno = math.cos(radianes)
tangente = math.tan(radianes)

arcSeno = math.asin(seno)
arcCoseno = math.acos(coseno)
arcTangente = math.atan(tangente)

print(f'{grados}°\t\t\t\t= {radianes} rads')
print(f'sen({radianes})\t\t= {seno}')
print(f'cos({radianes})\t\t= {coseno}')
print(f'tan({radianes})\t\t= {tangente}')
print(f'arcSen({seno})\t= {arcSeno} rads = {arcSeno*180/math.pi}°')
print(f'arcCos({coseno})\t= {arcCoseno} rads = {arcSeno*180/math.pi}°')
print(f'arcTan({tangente})\t= {arcTangente} rads = {arcSeno*180/math.pi}°')
```

Python permite dar un formato requerido a la salida de los cálculos.

```python
import math

grados = float(input('Ingresa un ángulo en grados '))

radianes = grados * math.pi / 180

seno = math.sin(radianes)
coseno = math.cos(radianes)
tangente = math.tan(radianes)

arcSeno = math.asin(seno)
arcCoseno = math.acos(coseno)
arcTangente = math.atan(tangente)

print(f'{grados:.4f}°\t= {radianes:.4f} rads')
print(f'sen({radianes:.4f})\t= {seno:.4f}')
print(f'cos({radianes:.4f})\t= {coseno:.4f}')
print(f'tan({radianes:.4f})\t= {tangente:.4f}')
print(f'arcSen({seno:.4f})\t= {arcSeno:.4f} rads = {arcSeno*180/math.pi:.4f}°')
print(f'arcCos({coseno:.4f})\t= {arcCoseno:.4f} rads = {arcCoseno*180/math.pi:.4f}°')
print(f'arcTan({tangente:.4f})\t= {arcTangente:.4f} rads = {arcTangente*180/math.pi:.4f}°')
```
